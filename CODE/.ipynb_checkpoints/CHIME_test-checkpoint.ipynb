```python
import RMextract.getRM as gt
from astropy.time import Time
import numpy as np
from astropy import units as u
from astropy.coordinates import Angle
from astropy.coordinates import EarthLocation
import os 
import IFR_subroutines as subs
import importlib as imp
import matplotlib.pyplot as plt
import csv
```

Initial tests for plotting ionospheric RM based on RMextract for the CHIME Q and U maps. May be outdated. Check for better version on Cedar. AO 31 Aug. 2021

```python
loc = EarthLocation(lat="49:19:17.4", lon="-119:37:12", height=542.2*u.m)
print(loc.get_itrs)

# as calculated by ALBUS based on the same coordinates:
# [-2058896.67046392 -3621369.67824459  4814351.45421717]
# seems like it's using the same function.
```

```python
with open('../2235_dec_test.csv', newline='') as csvfile:
    dec = list(csv.reader(csvfile))
dec = np.asarray(dec)[:,0].astype(float)
print(dec.shape)

with open('../2235_ra_test.csv', newline='') as csvfile:
    ra = list(csv.reader(csvfile))
ra = np.asarray(ra)[:,0].astype(float)
print(ra.shape)
```

```python
t = Time(['2019-12-23T01:54:12'],format='isot',scale ='utc')
starttime = t.mjd*24*3600.  # getRM still wants MJD time in seconds (casacore definition)
#endtime = starttime + 900
statpos = [loc.itrs.x.value, loc.itrs.y.value, loc.itrs.z.value]
pointing=[ 0.0,  90.*np.pi/180. ]

RMdict = gt.getRM(ionexPath='./IONEXdata/', radec=pointing, timestep=900, timerange = [starttime, starttime], stat_positions=[statpos,])

times1=RMdict['times']
RM1 = RMdict['RM']['st1']


time = Time(times1[1]/(24*3600),format='mjd')
time2 = time.isot
print(time2)
#print(Time(times1,scale='utc',format='mjd').,RM1)
```

```python
pixratio = 16

xpix = int(len(ra)/pixratio)
ypix = int(len(dec)/pixratio)

print(xpix,ypix)

dt = 23.9344696/xpix
t0 = Time(['2019-12-23T01:54:12'],format='isot',scale ='utc')+(dt/2)*u.hour
t_arr = []

dec_arr = np.empty([ypix])
ra_arr = np.empty([xpix])

for i in range(0,xpix):
    t_arr.append(t0+i*dt*u.hour)
    ra_arr[i] = np.mean(ra[i*pixratio:(i+1)*pixratio])
    #print(i*pixratio,(i+1)*pixratio)

for i in range(0,ypix):
    dec_arr[i] = np.mean(dec[i*pixratio:(i+1)*pixratio])
```

```python
for i in range(0,xpix):
    print(t_arr[i],ra_arr[i])

print(dec_arr)
```

```python
plt.scatter(ra,ra)
plt.scatter(ra_arr,ra_arr)
plt.xlim(0,60)
plt.ylim(0,60)
```

```python
plt.scatter(dec,dec)
plt.scatter(dec_arr,dec_arr)
#plt.xlim(0,30)
#plt.ylim(0,30)
```

```python
statpos = [loc.itrs.x.value, loc.itrs.y.value, loc.itrs.z.value]

RMs = np.empty([ypix,xpix])

for i in range(0,xpix):
    print('=============')
    print(i)
    print('=============')
    starttime = t_arr[i].mjd*24*3600.
    for j in range(0,ypix):
        print(j)
        pointing = [ra_arr[i]*np.pi/180., dec_arr[j]*np.pi/180.] #Ra, Dec in radians
        RMdict = gt.getRM(ionexPath='./IONEXdata/', radec=pointing, timestep=dt, 
                          timerange = [starttime, starttime], stat_positions=[statpos,])
        #print(RMdict['RM']['st1'])
        #time = Time(RMdict['times']/(24*3600),format='mjd')
        #time2 = time.isot
        #print(time2)
        RMs[j,i] = RMdict['RM']['st1'][1][0]

#print(RMs)
```

```python
plt.imshow(RMs)
```

```python
fig, ax = plt.subplots(figsize=(12, 4))
vmin=0.0
vmax=1.0
cmap='plasma'
im = ax.imshow(np.flip(RMs,[0,1])[6:22,:], origin='lower', aspect='auto',
                   extent=(ra_arr[-1], ra_arr[0], dec_arr[6], dec_arr[22]),
                   vmin=vmin, vmax=vmax, cmap=cmap)
ax.plot([210,210],[-20,100],linestyle='dashed',color='black')
ax.plot([330,330],[-20,100],linestyle='dashed',color='black')
#declabel = [r'${:.1f}^\circ$'.format(d) for d in np.degrees(np.arcsin(ax.get_yticks())) + eph.CHIMELATITUDE]
ax.set_ylim(dec_arr[6], dec_arr[22])
#rax = ax.twinx()
#rax.set_ylim(ax.get_ylim())
#rax.set_yticklabels(declabel)
#rax.set_ylabel('Declination (degrees)')

#ax.axvline(x=eph.lsa(eph.solar_rising(mapdate)), linestyle=':', color='k')
#ax.axvline(x=eph.lsa(eph.solar_setting(mapdate)),linestyle=':', color='k')
#ax.xaxis.set_major_locator(MultipleLocator(60))
ax.set_xlabel('Right ascension')
ax.set_ylabel('Declination')
ax.set_title('Ionosphere Dec. 23 2019')
cb = fig.colorbar(im)
cb.set_label('rad m$^{-2}$')
plt.savefig('../PLOTS/iono_CHIME_Dec23_2019.pdf')
```
