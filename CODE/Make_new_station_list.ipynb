```python
f = open('nrcan_stationlist_Apr19.csv')

larr = np.empty(numline)
barr = np.empty(numline)

for i in range(0,numline):
    line = f.readline()
    if i == 0 :
        print(line)
        print('')
    else:
        linesplit = line.split()
        if linesplit[5].split()[0] == str('-'):
            ldec = Angle(360.-np.float(linesplit[6].split(',')[5]), u.degree)
            bdec = Angle(np.float(linesplit[6].split(',')[3]), u.degree)
            h = str(np.round(np.float(linesplit[6].split(',')[7]),1)).rjust(8, ' ')
            larr[i] = 360.-np.float(linesplit[6].split(',')[5])
            barr[i] = np.float(linesplit[6].split(',')[3])
        else:
            ldec = Angle(360.-np.float(linesplit[5].split(',')[5]), u.degree)
            bdec = Angle(np.float(linesplit[5].split(',')[3]), u.degree)
            h = str(np.round(np.float(linesplit[5].split(',')[7]),1)).rjust(8, ' ')
            larr[i] = 360.-np.float(linesplit[5].split(',')[5])
            barr[i] = np.float(linesplit[5].split(',')[3])
        lon = ldec.to_string(unit=u.degree, sep=' ',precision=1)
        lat = bdec.to_string(unit=u.degree, sep=' ',precision=1)
        print(' '+str.lower(linesplit[0][-4:])+'  A XXXXXXXXX P                        '+lon+'  '+lat+h)
        
        #print(h)
        #print('')
```

```python
f = open('Washington_gnss.csv')
for i in range(0,numline):
    line = f.readline()
    if i < 2 :
        print(line)
    else:
        linesplit = line.split()

        ldec = Angle(360.+np.float(linesplit[0].split(',')[7]), u.degree)
        bdec = Angle(np.float(linesplit[0].split(',')[6]), u.degree)
        h = str(np.round(np.float(linesplit[0].split(',')[8]),1)).rjust(8, ' ')

        lon = ldec.to_string(unit=u.degree, sep=' ',precision=1)
        lat = bdec.to_string(unit=u.degree, sep=' ',precision=1)
        print(' '+str.lower(linesplit[0].split(',')[2])+'  A XXXXXXXXX P                        '+lon+'  '+lat+h)
```

```python
f = open('../DATA/gps_pos_default_v2.snx')
numline = 0
for line in f:
    numline = numline+1
print(numline)

all_stn = []
all_lon = np.empty(numline-20)
all_lat = np.empty(numline-20)

f = open('../DATA/gps_pos_default_v2.snx')
for i in range(0,numline-1):
    line = f.readline()
    if i >18:
        all_stn.append(str(line.split()[0]))
        all_lon[i-19] = float(line.split()[-7])+float(line.split()[-6])/60.+float(line.split()[-5])/3600.
        all_lat[i-19] = float(line.split()[-4])+float(line.split()[-3])/60.+float(line.split()[-2])/3600.
        
print(len(all_stn))
print(len(all_lon))

all_stn = np.asarray(all_stn)

all_lon[all_lon>180.] = all_lon[all_lon>180.]-360.

for i in range(0,len(all_stn)):
    print(all_stn[i],all_lon[i],all_lat[i])
```

```python
good_stn = []

f = open('../../IFR_tests/DRAO26_4/albus_good_stations')
numline = 0
for line in f:
    numline = numline+1
print(numline)
f = open('../../IFR_tests/DRAO26_4/albus_good_stations')
for i in range(0,numline):
    line = f.readline(6).split()[0]
    good_stn.append(line)

good_stn = np.asarray(good_stn)    

good_lon = np.empty(len(good_stn))
good_lat = np.empty(len(good_stn))

for i in range(0,len(good_stn)):
    w = np.where(all_stn == good_stn[i])
    good_lon[i] = all_lon[w[0][0]]
    good_lat[i] = all_lat[w[0][0]]
    #print(good_stn[i],w[0][0])
    #print('')
 
```

```python
f = open('../../IFR_tests/DRAO26_4/albus_potential_stations')
numline = 0
for line in f:
    numline = numline+1
print(numline)

ptl_stn = []
f = open('../../IFR_tests/DRAO26_4/albus_potential_stations')
for i in range(0,numline):
    line = f.readline().split()[3]
    #print(line)
    ptl_stn.append(line)
print(numline)

ptl_stn = np.asarray(ptl_stn)
#print(ptl_stn)

bad_stn = []
for i in range(0,numline):
    w = np.where(good_stn == ptl_stn[i])
    #print(len(w[0]))
    if len(w[0]) == 0:
        bad_stn.append(ptl_stn[i])

bad_stn = np.asarray(bad_stn)
        
print(len(bad_stn))
print(bad_stn)

bad_lon = np.empty(len(bad_stn))
bad_lat = np.empty(len(bad_stn))

for i in range(0,len(bad_stn)):
    w = np.where(all_stn == bad_stn[i])
    bad_lon[i] = all_lon[w[0][0]]
    bad_lat[i] = all_lat[w[0][0]]
```

```python
m = folium.Map(location=[49.3, -119.6],tiles="Stamen Terrain")

good_pts = len(good_lon)
bad_pts = len(bad_lon)

print(good_pts)
print(bad_pts)

for i in range(0,good_pts):
    folium.CircleMarker(location=[good_lat[i],good_lon[i]],popup=good_stn[i], 
                       radius = 3, color='blue').add_to(m)
for i in range(0,bad_pts):
    folium.CircleMarker(location=[bad_lat[i],bad_lon[i]],popup=bad_stn[i],
                       radius = 2, color='red').add_to(m)


m
```
